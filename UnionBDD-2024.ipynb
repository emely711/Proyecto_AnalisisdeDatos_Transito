```python
import unicodedata
import pandas as pd


def normaliza(s):
    """Normaliza texto removiendo tildes y caracteres especiales."""
    if not isinstance(s, str):
        return ""
    s = unicodedata.normalize("NFKD", s)
    return s.encode("ascii", "ignore").decode("ascii").lower().strip()


def seleccionar_40_columnas(columnas_totales):
    """Selecciona con prioridad alta las columnas que coincidan con tus variables

    clave y completa exactamente hasta 40 columnas.
    """
    palabras_clave = [
        "placa",
        "vehiculo",
        "accidente",
        "siniestro",
        "hora",
        "fecha",
        "causa",
        "servicio",
        "sexo",
        "genero",
        "edad",
        "nombre",
        "provincia",
        "canton",
    ]

    prioritarias = []
    secundarias = []

    for col in columnas_totales:
        col_norm = normaliza(str(col))
        if any(kw in col_norm for kw in palabras_clave):
            prioritarias.append(col)
        else:
            secundarias.append(col)

    # Combinar prioritarias + secundarias y limitar a 40
    seleccionadas = prioritarias + secundarias
    return seleccionadas[:40]


# ---------------------------------------------------------
# PROCESAMIENTO DEL ARCHIVO 2024 (AGOSTO - DICIEMBRE)
# ---------------------------------------------------------
archivo_entrada = "BDD_2024_AGOSTO_DICIEMBRE.xlsx"
archivo_salida = "BDD_2024_AGOSTO_DICIEMBRE_40COLS.xlsx"

print(f"Procesando archivo 2024: {archivo_entrada}...")

# 1. Inspeccionar nombres de columnas en la primera fila
preview = pd.read_excel(archivo_entrada, nrows=1)

# 2. Elegir las 40 columnas prioritarias
cols_40 = seleccionar_40_columnas(preview.columns)

# 3. Cargar el DataFrame solo con esas 40 columnas
df_2024 = pd.read_excel(archivo_entrada, usecols=cols_40)

# 4. Guardar el resultado en Excel
df_2024.to_excel(archivo_salida, index=False)

print(
    f"✓ ¡Listo! Guardado con éxito: {archivo_salida} ({df_2024.shape[0]} filas x {df_2024.shape[1]} columnas)"
)
```
